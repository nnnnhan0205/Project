```python
#import libary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Understand data

```python
df1 = pd.read_csv('car_detail_en.csv')
```

```python
df2 = pd.read_csv('seller_en.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.shape
```

```python
df2.shape
```

```python
#detele unused column
df1.drop(['url', 'fuel_consumption','fuel_system'], axis = 1, inplace=True)
df2.drop(['Phone 2', 'Website',], axis = 1, inplace=True)
```

```python
print(df1.shape, df2.shape)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, how="left", on=["ad_id"]) #merge df1 and df2 to df
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(how='all', inplace=True) #drop null value
```

```python
df = df.drop_duplicates() #drop dulicated
```

```python
#covert str value to numric in price column, add new column
def convert_to_numeric(value):
    try:
        if 'Billion' in value and not 'Million' in value:
            return float(value.replace('Billion', '').replace(' ', '')) * 1e3
        elif 'Million' in value and not 'Billion' in value:
            return float(value.replace('Million', '').replace(' ', ''))
        else:
            return float(value.replace('Million', '').replace('Billion', '').replace(' ', ''))
    except:
        return None
df['log10_price'] = np.log10(df['price. price'].apply(convert_to_numeric))
```

```python
df.head(10)
```

# Visualization

```python
fig, ax = plt.subplots(figsize = (12,8))
count_brand = df.groupby('brand')['brand'].count().sort_values(ascending=False )
count_brand.plot(kind='bar')
plt.title('Count of Brand')
plt.xlabel('Brand')
plt.ylabel('Quantity')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
count_origin = df.groupby('origin')['origin'].count()
count_origin.plot(kind='bar')
plt.title('Quantity of origin')
plt.xlabel('origin')
plt.ylabel('Quantity')
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
count_num_of_doors = df.groupby('num_of_doors')['num_of_doors'].count().sort_values(ascending=False )
count_num_of_doors.plot(kind='bar')
plt.title('Quantity of num_of_doors')
plt.xlabel('num_of_doors')
plt.ylabel('Quantity')
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()

fig, ax = plt.subplots(figsize = (12,8))
count_num_of_doors = df.groupby('seating_capacity')['seating_capacity'].count().sort_values(ascending=False )
count_num_of_doors.plot(kind='bar')
plt.title('Quantity of seating_capacity')
plt.xlabel('seating_capacity')
plt.ylabel('Quantity')
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
count_car_model = df.groupby('car_model')['car_model'].count().sort_values(ascending=False )
count_car_model.plot(kind='bar')
plt.title('Quantity of car_model')
plt.xlabel('car_model')
plt.ylabel('Quantity')
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
count_exterior_color = df.groupby('exterior_color')['exterior_color'].count().sort_values(ascending=False )
count_exterior_color.plot(kind='bar')
plt.title('Quantity of exterior_color')
plt.xlabel('exterior_color')
plt.ylabel('Quantity')
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()

fig, ax = plt.subplots(figsize = (12,8))
count_interior_color = df.groupby('interior_color')['interior_color'].count().sort_values(ascending=False )
count_interior_color.plot(kind='bar')
plt.title('Quantity of interior_color')
plt.xlabel('interior_color')
plt.ylabel('Quantity')
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
count_condition = df.groupby('condition')['condition'].count()
count_condition.plot(kind='pie',labeldistance=1.1,autopct = '%1.1f%%')
plt.title('Quantity of condition')
textprops = {"fontsize":12}
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
sum_brand = df.groupby(['brand'])['log10_price'].sum().sort_values(ascending=False ).head(10)
sum_brand.plot(kind='bar')
plt.title('Top 10 Revenue of Brand')
plt.xlabel('Brand')
plt.ylabel('Revenue')
ax.bar_label(ax.containers[0], label_type='edge')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
sum_car_name = df.groupby(['car_name'])['log10_price'].sum().sort_values(ascending=False ).head(10)
sum_car_name.plot(kind='bar')
plt.title('Top 10 Revenue of Car Model')
plt.xlabel('car_name')
plt.ylabel('Revenue')
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```
