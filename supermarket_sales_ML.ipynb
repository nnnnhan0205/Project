```python
# revenue of city
fig, ax = plt.subplots(figsize=(12, 8))
city_revenue = df.groupby('city')['revenue'].sum()
city_revenue.plot(kind='bar')
plt.title('Revenue of City')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
#revenue of branch
fig, ax = plt.subplots(figsize=(12, 8))
branch_revenue = df.groupby('branch')['revenue'].sum()
branch_revenue.plot(kind='bar')
plt.title('Revenue of Branch')
plt.xlabel('Branch')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
#revenue of customer type
fig, ax = plt.subplots(figsize=(12, 8))
customer_type_revenue = df.groupby('customer_type')['revenue'].sum()
customer_type_revenue.plot(kind='bar')
plt.title('Revenue of Customer Type')
plt.xlabel('customer_type')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
#revenue of product line
fig, ax = plt.subplots(figsize=(12, 8))
product_line_revenue = df.groupby('product_line')['revenue'].sum()
product_line_revenue.plot(kind='bar')
plt.title('Revenue of Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
#Quantity Of Payment Method
fig, ax = plt.subplots(figsize=(10, 6))
payment_method_quantity = df.groupby('payment_method')['quantity'].sum()
payment_method_quantity.plot(kind='bar')
plt.title('Quantity Of Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('quantity')
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
#Sum Of Rating Branch
fig, ax = plt.subplots(figsize=(12, 8))
Sum_of_rating_branch = df.groupby('branch')['rating'].mean()
Sum_of_rating_branch.plot(kind='bar')
plt.title('Sum Of Rating Branch')
plt.xlabel('Branch')
plt.ylabel('Rating')
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], label_type='edge')
plt.show()
```

```python
#Gross Income of City
plt.figure(figsize=(12, 8))
city_gross_income =df.groupby('city')['gross_income'].sum()
city_gross_income.plot(kind='pie',labeldistance=0.98,autopct = '%1.1f%%')
textprops = {"fontsize":12}
plt.title('Gross Income of City')
plt.show()
#Gross Income of Branch
plt.figure(figsize=(12, 8))
branch_gross_income =df.groupby('branch')['gross_income'].sum()
branch_gross_income.plot(kind='pie',autopct = '%1.1f%%')
textprops = {"fontsize":12}
plt.title('Gross Income of branch')
plt.show()
#Gross Income of Customer Type
plt.figure(figsize=(12, 8))
customer_type_gross_income =df.groupby('customer_type')['gross_income'].sum()
customer_type_gross_income.plot(kind='pie',autopct = '%1.1f%%')
textprops = {"fontsize":12}
plt.title('Gross Income of Customer Type')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='product_line', data=df, palette = "viridis")
plt.title('Distribution of Product Lines')
plt.show()
```

```python
#Box plot of Gross Income by City
plt.figure(figsize=(12, 8))
sns.boxplot(x='city', y='gross_income', data=df,palette = "viridis")
plt.title('Boxplot of Gross Income by City')
plt.show()
```

```python
#heatmap
numeric_columns = df.select_dtypes(include=[np.number]).columns
numeric_df = df[numeric_columns]
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()
```

Build ML moldel

```python
features = df[['unit_cost','rating','quantity']]
X = features
target = df[['gross_income']]
y = target
```

```python
# Split the data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('Model Evaluation:')
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R-squared:', r2_score(y_test, y_pred))
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state = 1)
```

```python
dt_model.fit(X_train,y_train)
```

```python
y_pred2 = dt_model.predict(X_test)
y_pred2
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('Model Evaluation:')
print('Mean Squared Error:', mean_squared_error(y_test, y_pred2))
print('R-squared:', r2_score(y_test, y_pred2))
```
